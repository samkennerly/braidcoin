```python
import networkx as nx
import matplotlib.pyplot as plt

from minernet import *
```

```python
FIGSIZE = (9,3)
```

```python
beads = [
    Bead(0),
    Bead(1, parents=(0,)),
    Bead(2, parents=(0,)),
    Bead(3, parents=(0,)),
    Bead(4, parents=(1,2)),
    Bead(5, parents=(4,3)),
    Bead(6, parents=(5,)),
    Bead(7, parents=(6,)),
    Bead(8, parents=(6,)),
    Bead(9, parents=(7,8))
]
beads
```

```python
links = [ (y, x.name) for x in beads for y in x.parents]
links
```

```python
braid = nx.DiGraph()
braid.add_edges_from(links)
braid
```

```python
plt.figure(figsize=FIGSIZE)
nx.draw_networkx(braid)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.kamada_kawai_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.shell_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.planar_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.bfs_layout(braid, start=list(braid.nodes)[0])
nx.draw_networkx(braid, pos=pos)
```
