```python
import networkx as nx
import matplotlib.pyplot as plt

from minernet import *
```

```python
beads = [
    Bead(0),
    Bead(1, parents=(0,)),
    Bead(2, parents=(0,)),
    Bead(3, parents=(0,)),
    Bead(4, parents=(1,2)),
    Bead(5, parents=(3,)),
    Bead(6, parents=(4,5)),
    Bead(7, parents=(6,)),
    Bead(8, parents=(6,)),
    Bead(9, parents=(7,8)),
    Bead(10, parents=(9,)),
]
braid = Braid(beads)
braid
```

```python
links = braid.links
root = braid.root

graph = DiGraph()
graph.add_edges_from(links())

pos = bfs_layout(graph, start=root.name)
fig = figure(figsize=(9,3))
draw_networkx(graph, pos=pos)
```
