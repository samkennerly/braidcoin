```python
import networkx as nx
import matplotlib.pyplot as plt

from minernet import *
```

```python
FIGSIZE = (9,3)
```

```python
beads = [
    Bead(0),
    Bead(uid=1, parents=(0,)),
    Bead(uid=2, parents=(0,)),
    Bead(uid=3, parents=(0,)),
    Bead(uid=4, parents=(1,2)),
    Bead(uid=5, parents=(4,3)),
]
beads
```

```python
links = [ (x.uid,y) for x in beads for y in x.parents]
links
```

```python
braid = nx.DiGraph()
braid.add_edges_from(links)
braid
```

```python
plt.figure(figsize=FIGSIZE)
nx.draw_networkx(braid)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.kamada_kawai_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.shell_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.planar_layout(braid)
nx.draw_networkx(braid, pos=pos)
```

```python
fig = plt.figure(figsize=FIGSIZE)
pos = nx.bfs_layout(braid, start=5)
nx.draw_networkx(braid, pos=pos)
```
